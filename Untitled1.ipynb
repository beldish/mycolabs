```python
# https://www.youtube.com/watch?v=s2K1JfNR7Sc&t=144s
#------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16,12]
plt.rcParams.update({'font.size':18})

# Create a simple signal with two frequencies
dt = 0.001
t = np.arange(0,1,dt)
f = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t)  # Sum of two frequencies
f_clean = f
f = f + 2.5* np.random.randn(len(t))              # add some noise

ax = plt.axes()
ax.set_facecolor("black")

plt.plot(t,f,color='red', label='Noisy')
plt.plot(t,f_clean, color='white', label='Clean')
plt.xlim(t[0], t[-1])
plt.legend()
```

```python
## Compute the Fats Fourier Transfor (FFT)
n = len(t)
fhat = np.fft.fft(f, n)                       # compute the FFT
PSD = fhat *np.conj(fhat) / n                 # Power spectrum
freq = ( 1/ (dt*n)) * np.arange(n)            # Create X axis of frequencies
L = np.arange(1, np.floor(n/2), dtype='int')

fig, axs = plt.subplots(2,1)
axs[0].set_facecolor("black")
axs[1].set_facecolor("black")

plt.sca(axs[0])
plt.plot(t,f, color='red', label='Noisy')
plt.plot(t,f_clean, color='white', label='Clean')
plt.xlim(t[0], t[-1])
plt.legend()

plt.sca(axs[1])
plt.plot(freq[L], PSD[L], color='red', label='Noisy')
plt.xlim(freq[L[0]], freq[L[-1]])
plt.legend()

plt.show()
```
